## Dataset Overview

> This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('clean_data.csv')
```

## What is the duration distribution ?
>as the next plot well indicates the majority of the duration are around 3 to 15 minutes


```python
plt.figure(figsize=(10,5))
bins = np.arange(1, df.duration_min.max(),2)
plt.hist(data=df , x = 'duration_min',bins = bins );
plt.xlabel('duration')
plt.ylabel('number of members');
plt.title('duration distribution');
```

## what is days with the lowest duration ?

> as the next plot will indicates the duration in saterday and sunday is the lowest among the whole week

```python
order = ['Saturday' , 'Sunday' , 'Monday' , 'Tuesday' , 'Wednesday' , 'Thursday', 'Friday']
g = sb.FacetGrid(data = df , col='start_day' , col_wrap=4 , col_order=order  )
g.map(plt.hist , 'duration_min' );
g.set_ylabels('Duration')
g.fig.subplots_adjust(top=0.88)
g.fig.suptitle('Total duration per day');
```

## What is the age distribution of the users

> as the next plot will indicates the majority of the users are in after their 25 to 35

```python
plt.figure(figsize=(10,5))
bins = np.arange(20 , df.age.max(),2)
plt.hist(data=df , x = 'age',bins = bins );
plt.xlabel('Age')
plt.ylabel('number of members');
plt.title('Age distribution');
```
